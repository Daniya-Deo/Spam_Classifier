```python
import string
def cleanText(message):
    message = re.sub('[^a-zA-Z]', ' ', message)
    message = message.lower()
    message = message.split()
    words = [ss.stem(word) for word in message if word not in stop]
    return " ".join(words)

msg_df["message"] = msg_df["message"].apply(cleanText)
msg_df.head(n = 10)

```

```python
spam_messages = msg_df[msg_df["label"] == "spam"]["message"]
ham_messages = msg_df[msg_df["label"] == "ham"]["message"]
```

```python
spam_messages
```

```python
spam_words = []
ham_words = []

def extractSpamWords(spamMessages):
    global spam_words
    words = [word for word in word_tokenize(spamMessages)]
    spam_words = spam_words + words

def extractHamWords(hamMessages):
    global ham_words
    words = [word for word in word_tokenize(hamMessages) ]
    ham_words = ham_words + words

spam_messages.apply(extractSpamWords)
ham_messages.apply(extractHamWords)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Assuming msg_df is your DataFrame containing the 'message' and 'length' columns
# Convert the 'message' column to a list of strings
messages = msg_df['message'].tolist()

# Create the CountVectorizer object and fit_transform the messages
cv = CountVectorizer()
X = cv.fit_transform(messages)

# Convert the sparse matrix to a dense matrix and create the DataFrame
df = pd.DataFrame(X.toarray(), columns=cv.get_feature_names_out())

# Add the 'length' column to the DataFrame
df['len'] = msg_df['length']

# Now you have the DataFrame 'df' with the word counts as columns and the 'length' column
print(df)
```

```python
from sklearn.model_selection import train_test_split

y=msg_df['label']

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.20, random_state = 0)

# Training model using Naive bayes classifier

from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(fbeta_score(y_test,y_pred,beta =0.5))
```
